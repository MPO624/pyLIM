# Analyzing local anomaly correlations...

```python
import pandas as pd
import numpy as np
from IPython.display import Image
%load_ext autoreload
%autoreload 1
%aimport FileTools
```

```python
corr = pd.read_hdf('fcast_corr.h5', 'w')
lats = np.load('latgrd.npy')
lons = np.load('longrd.npy')
```

```python
#Test on single forecast time
data = corr[24:25].values#~2 year forecast
title = 'Local Anomaly Correlation: 2-year forecast'
filename = 'lca_2tst.png'
FileTools.plot_corrdata(lats, lons, data.reshape(lons.shape), title, filename)
Image(filename=filename)
```

```python
#2-5 year forecasts
data = corr[24:61] #~2-5 year forecast
mean_corr = data.mean(axis=0).values
title = 'Local Anomaly Correlation: 2-5 year forecast'
filename = 'lca_2_5yr.png'
FileTools.plot_corrdata(lats, lons, mean_corr.reshape(lons.shape), title, filename)
Image(filename=filename)
```

```python
print 'Max Correlation:',mean_corr.max()
print 'Min Correlation:',mean_corr.min()
```
